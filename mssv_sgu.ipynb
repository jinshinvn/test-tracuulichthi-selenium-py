Lọc mã sinh viên trùng nhau từ sample-2.xlsx sang data-mssv-2-exported.csv

```python
import pandas as pd
df = pd.ExcelFile('data/sample-2.xlsx').parse('DSSV_PhongThi')
```

Lấy mã sinh viên, họ và tên, lớp thôi. Mấy cái khác bỏ đêi :D

```python
df_res = df.loc[:,["Mã SV", "Họ và tên", "Lớp"]].drop_duplicates(subset="Mã SV")
```

Export ra thôi

```python
df_res.to_csv('data-mssv-2-exported.csv')
```
